# EXPLORATORY DATA ANALYSIS

```python
# Histogram plot for distribution of total sales:
plt.figure(figsize=(10,4))
sns.histplot(df['Total'],bins=30,color='black',edgecolor='Black',kde='True')
plt.title('Distribution of Total Sales')
plt.show()
```

**The histogram shows the distribution of total sales across all product and branches. The distribution appears to be right-skewed, with most frequent total sales are between 100-600**

```python
# Histogram plot for distribution of rating:
plt.figure(figsize=(10,4))
sns.histplot(df['Rating'],bins=30,color='black',edgecolor='black',kde=True)
plt.title('Distribution of Rating')
plt.show()
```

```python
# value counts of city:
city_count=df['City'].value_counts()
print(city_count)

# Bar plot for count of city:
plt.figure(figsize=(10,4))
plt.bar(city_count.index,city_count.values,color=sns.color_palette(),edgecolor='black')
plt.title('Distribution of Cities')
plt.xlabel('City')
plt.ylabel('Counts')
plt.show()
```

**The bar plot shows the most distribution of city. Yangon is a highest number of cities to sale the product.**

```python
# values counts of customer type:
customer_type_counts=df['Customer type'].value_counts()
print(customer_type_counts)

# Bar plot for customer type:
plt.figure(figsize=(10,4))
plt.bar(customer_type_counts.index,customer_type_counts.values,color=sns.color_palette(),edgecolor='black')
plt.title('Distribution of customer type')
plt.xlabel('Customer type')
plt.ylabel('Counts')
plt.show()
```

```python
# value counts of product line:
product_line_counts=df['Product line'].value_counts()
print(product_line_counts)

# Bar plot for product line:
plt.figure(figsize=(10,4))
plt.bar(product_line_counts.index,product_line_counts.values,color=sns.color_palette(),edgecolor='black')
plt.title('Distribution of Product line')
plt.xticks(rotation=45)
plt.xlabel('Product Line')
plt.ylabel('Counts')
plt.show()
```

**The bar plot shows the distribution of product line across all product. The Fashion accessories is most counts of product in product line.**

```python
genderwise_total_sales=df.groupby('Gender')['Total'].mean()
print(genderwise_total_sales)

# Pie plot for Total by Gender
plt.figure(figsize=(10,4))
plt.pie(x=genderwise_total_sales,labels=genderwise_total_sales.index,colors=sns.color_palette(),startangle=45,autopct='%1.1f%%',explode=[0.05]*len(genderwise_total_sales))
plt.title('Total Sales by Gender')
plt.axis('equal')
plt.show()
```

**The pie chart shows total sales based on genders. Female(51.9%) are with highest percentages of purchasing the product when compare to male(48.1%).**

```python
# scatter plot the relation unit price and total:
plt.figure(figsize=(10,4))
sns.scatterplot(x='Unit price',y='Total',data=df,color='blue')
plt.title('Unit Price vs Total sales')
plt.show()
```

```python
plt.figure(figsize=(18,5))

plt.subplot(141)
df.groupby('City')['Total'].mean().plot(kind='pie',autopct='%1.1f%%',startangle=200)
plt.title('1.sales by city')

plt.subplot(142)
df.groupby('Payment')['Total'].mean().plot(kind='pie',autopct='%1.1f%%',startangle=150)
plt.title('2.Payment Mode for Total sales')

plt.subplot(143)
df.groupby('Branch')['Total'].mean().plot(kind='pie',autopct='%1.1f%%',startangle=100)
plt.title('3.Branch-wise Total Sales')

plt.tight_layout()
plt.show()
```

**The pie chart used to comparison of Average Sales vs city, payment mode and branch.**

**Average sale vs City**
> The Naypyitaw has a highest sales when compare to other cities.

**Average sale vs Payment Mode**
> The Most gender used the cash as a payment mode for bought the product line.

**Average sale vs Branches**
> The Most sales on the branch wise is C branch.

```python
# Boxplot for total sales by product line
plt.figure(figsize=(10,6))
sns.boxplot(x='Product line', y='Total', data=df)
plt.title('Total Sales by Product Line')
plt.xticks(rotation=45)
plt.show()
```

**The Boxplot shows the highest sales of product line. Home and lifestyle is maximum sales product line in the supermarket store.**

```python
# pie plot for branch vs Total sales and rating
plt.figure(figsize=(19,5))

plt.subplot(141)
df.groupby('Product line')['Rating'].mean().plot(kind='pie',autopct='%1.1f%%')
plt.title('Average Rating for Product line ')

plt.subplot(142)
df.groupby('Branch')['Rating'].mean().plot(kind='pie',autopct='%1.1f%%')
plt.title('Branch-wise Rating')

plt.tight_layout()
plt.show()
```

**Average rating vs Product line**
>The people gave highest average rating for the Food and beverages (17.0%).

**Branch vs Rating**
>The C branch gets the highest rating in the branch wise supermarket store (33.8%).


```python
# Select only numerical value for pairplot
numerical_columns=('Unit price','Quantity','Tax 5%','Total','Rating')

# Pairplot for selected Numerical features
sns.pairplot(df,hue='Gender',vars=numerical_columns,diag_kind='kde',height=3.5,aspect=1.2)
plt.suptitle('Pairplot with Custom Option',y=1.02)
plt.show()
```
